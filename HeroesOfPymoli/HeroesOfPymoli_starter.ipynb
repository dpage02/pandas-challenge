```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
numbers_purchased = purchase_data["Item ID"].count()
avg_price = purchase_data["Price"].mean()
total_revenue = purchase_data["Price"].sum(axis=0)

pd.options.display.float_format = '${:,.2f}'.format

purchasing_analysis_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                                      "Average Prics":[avg_price],
                                      "Number of Purchases":[numbers_purchased],
                                      "Total Revenue":[total_revenue]})
purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
demo_df = purchase_data.groupby('Gender')['SN'].nunique()

perc_player = (demo_df / total_players) * 100

pd.options.display.float_format = '{:,.2f}%'.format
gender_demo_df = pd.DataFrame({"Total Count":demo_df,
                               "Percentage of Players": perc_player
                              },
                             index=["Male","Female","Other / Non-Disclosed"])
gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pd.options.display.float_format = '${:,.2f}'.format

purchase_demo = purchase_data.groupby('Gender')['Price']
demo_count = purchase_demo.count()
avg_demo_purchase = purchase_demo.mean()
tot_purchase = purchase_demo.sum()
avg_tot_purchase = tot_purchase/gender_demo_df["Total Count"]
gender_purchasing_df = pd.DataFrame({"Purchase Count":demo_count,
                                    "Average Purchase Price":avg_demo_purchase,
                                    "Total Purchase Values":tot_purchase,
                                    "Avg Total Purchase per Person":avg_tot_purchase})

gender_purchasing_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_df = purchase_data[["Age","SN"]]
age_df = age_df.drop_duplicates("SN")

age_bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14","15-19","20-24",
              "25-29","30-34","35-39","40+"]
age_df["Age Group"] = pd.cut(age_df["Age"],age_bins,labels=group_names)
grouped_age_demo = age_df.groupby(["Age Group"])


tot_players = age_df["Age"].count()

tot_count = grouped_age_demo["Age"].count()
perc_age_group = (tot_count / tot_players)*100

pd.options.display.float_format = '{:,.2f}%'.format

age_demo_df = pd.DataFrame({"Total Count":tot_count,
                           "Percentage of Players":perc_age_group
                           })


age_demo_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data[["SN","Price"]]

grouped_spenders = top_spenders.groupby("SN")["Price"]
grouped_spenders_count = grouped_spenders.count()
avg_spenders_purch = grouped_spenders.mean()
tot_spending = grouped_spenders.sum()


top_spenders_df = pd.DataFrame({"Purchase Count": grouped_spenders_count,
                               "Average Purchase Price": avg_spenders_purch,
                                "Total Purchase Value": tot_spending
                               })
top_spenders_df.sort_values(by=["Total Purchase Value"], ascending=False).head()                         
```
